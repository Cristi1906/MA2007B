```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import tadasets
import ripser
import persim
```

Datos Simulados

```python
#Primero demos un conjunto de puntos simulados en 3 grupos
X1 = np.random.multivariate_normal([0,0],np.array([[1,0],[0,1]]),size = 10)
X2 = np.random.multivariate_normal([7,0],np.array([[1,0],[0,0.2]]),size = 8)
X3 = np.random.multivariate_normal([0,11],np.array([[1,0],[0,1]]),size = 12)
X = np.concatenate((X1,X2,X3))

plt.plot(X[:,0],X[:,1],'r+')
plt.axis('equal')
plt.show()
```

Datos de tadasets

```python
np.random.seed(2879) #puedes cambiar de semilla aleatoria a voluntad
```

```python
datos_Circunferencia = tadasets.dsphere(d=1,n=100, r=1, noise=0.3) #esta es una esfera de dimensión d=1 en el plano, radio 1 y en la cual se tomaron "aleatoriamente" 1000 puntos.
datos_Circunferencia 
plt.scatter(datos_Circunferencia[:, 0], datos_Circunferencia[:, 1])
plt.axis('equal')
plt.show()
```

Diagramas de Persistencia

```python
def diagram_sizes(dgms):
    return ", ".join([f"|$H_{i}$|={len(d)}" for i, d in enumerate(dgms)])   #Aquí almacenamos la cantidad de generadores de cada espacio de homología. Estas cantidades se muestran en la impresión del mismo (puedes omitirlo si gustas, por supuesto haciendo los cambios necesarios en el código del diagrama)
```

Para los datos Simulados

```python
dgm_X = ripser.ripser(X,maxdim=2)['dgms'] #Aquí se construye y grafica el diagrama de persistencia (en este caso, de los puntos simulados). maxdim es la dimensión máxima en homología que quieres que se muestre.
persim.plot_diagrams(
    dgm_X, 
    show=True, 
    title=f"Datos Simulados\n{diagram_sizes(dgm_X)}"
) 

dgm_X = ripser.ripser(X,maxdim=1)['dgms'] #Aquí se construye y grafica el diagrama de persistencia (en este caso, de los puntos simulados). maxdim es la dimensión máxima en homología que quieres que se muestre.
persim.plot_diagrams(
    dgm_X, 
    show=True, 
    title=f"Datos Simulados\n{diagram_sizes(dgm_X)}"
)
```

Para la Circunferencia

```python
dgm_Circunferencia = ripser.ripser(datos_Circunferencia,maxdim=2)['dgms'] #Aquí se construye y grafica el diagrama de persistencia (en este caso, de los puntos de datos_Circunferencia). maxdim es la dimensión máxima en homología que quieres que se muestre.
persim.plot_diagrams(
    dgm_Circunferencia, 
    show=True, 
    title=f"Datos Circunferencia\n{diagram_sizes(dgm_Circunferencia)}"
) 
```

```python
#Ya vimos que para dimensión 2 no me da nada interesante lo dejo hasta dimensión 1
dgm_Circunferencia = ripser.ripser(datos_Circunferencia,maxdim=1)['dgms'] #Aquí se construye y grafica el diagrama de persistencia (en este caso, de los puntos de datos_Circunferencia). maxdim es la dimensión máxima en homología que quieres que se muestre.
persim.plot_diagrams(
    dgm_Circunferencia, 
    show=True, 
    title=f"Datos Circunferencia\n{diagram_sizes(dgm_Circunferencia)}"
) 
```
