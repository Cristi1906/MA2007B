```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data=np.linspace(0,2*np.pi,100)
plt.plot(data,np.cos(data),marker='o',color='blue')
cosdata=np.cos(data)
```

```python
def embed(points,window=2):
    auxembed=[]
    for i in range(0,len(points)-window+1):
        aux=[points[i+j] for j in range(window)]
        auxembed.append(aux)
    auxembed=np.array(auxembed)
    return auxembed

embededcos=embed(cosdata,window=2)
plt.plot(embededcos[:,0],embededcos[:,1],marker='o',color='blue')
```

```python
sindata=np.sin(data)
embededsin=embed(sindata,window=2)
plt.plot(embededsin[:,0],embededsin[:,1],marker='o',color='blue')
```

```python
nonperiodicdata=3*np.cos(3*data)+np.sin(data)
fig=plt.figure()
plt.plot(data,nonperiodicdata,marker='o',color='blue')
```

```python
embedednonperiodic=embed(nonperiodicdata,window=2)
plt.plot(embedednonperiodic[:,0],embedednonperiodic[:,1],marker='o',color='blue')
```

```python
import ripser
import persim
```

```python
ripserout=ripser.ripser(embedednonperiodic)['dgms']
```

```python
persim.plot_diagrams(
    ripserout, 
    show=True
) 
```

```python
ripscos=ripser.ripser(embededcos)['dgms']
persim.plot_diagrams(
    ripscos, 
    show=True
) 
```

```python
data=np.linspace(0,20*np.pi,1000)
datan=np.cos(2*data)+3*np.sin(5*data)
plt.plot(data,datan,marker='o',color='blue')
```

```python
def embed2(points,window=2,time_delay=100):
    auxembed=[]
    for i in range(0,len(points)+(-window+1)*time_delay):
        aux=[points[i+j*time_delay] for j in range(window)]
        auxembed.append(aux)
    auxembed=np.array(auxembed)
    return auxembed
```

```python
prueba=embed2(datan,window=2,time_delay=100)
z=np.zeros(len(prueba))
fig = go.Figure(data=[go.Scatter3d(x=prueba[:,0], y=prueba[:,1],z=z,
                                   mode='markers')])
fig.show()
```
