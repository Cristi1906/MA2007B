El objetivo de esta actividad es entender como funciona la preimagen de funciones entre espacios métricos.

Por ejemplo, consideremos la función $f:[0,2\pi]\to [-1,1]$ dada por $f(x)=\sin(x).$ Queremos saber si esta función es continua cuando consideramos la topología subespacio de $[0,2\pi]$ y $[-1,1]$ cuando se consideran subconjuntos de $(\mathbb{R},d_e).$

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

X = np.linspace(0,2*np.pi,1000)

Y = np.sin(x)


plt.scatter(X,Y)
```

```python
map={x:np.sin(x) for x in X}
f= lambda x: map[x]
# We wanto to find the preimage of an open in [-1,1]
def preimage(f, lim_inf,lim_sup, Domain): # Given a function, a subset A of the codomain and the Domain,
# the preimage function will return the preimage of A.
    aux = []
    # First we look for all preimage set of each element of A
    for x in Domain:
        if f(x) >= lim_inf and f(x) <= lim_sup:
            aux.append(x)
    # Finally we return the preimage set
    return aux
```

```python
Ainv=preimage(f,0,1,X)
```

```python
A=[f(x) for x in Ainv]
plt.scatter(Ainv,A)
```

Ahora graficaremos las preimagenes de una función de un subespacio de $\mathbb{R}^3$ en $\mathbb{R}.$

Aplicando lo que aprendiste en las secciones anteriores, gráfica algunas preimágenes para las funciones:
1. $f(x,y,z)=x.$
2. $f(x,y,z)=y.$
3. $f(x,y,z)=z.$

¿Estas preimágenes nos ayudan a conocer un poco mejor al subespacio?

¿Que pasa con las preimagenes donde parece que se traslapa el subespacio?

```python
def newtrefoil(t):
	#Parametrización del nudo trébol
    c, cc, ccc = np.cos(t), np.cos(2*t), np.cos(3*t) 
    s, ss, sss = np.sin(t), np.sin(2*t), np.sin(3*t) 
    x = 41*c-18*s-83*cc-83*ss-11*ccc+27*sss
    y = 36*c+27*s-113*cc+30*ss+11*ccc-27*sss
    z = 45*s-30*cc+113*ss-11*ccc+27*sss
    return [5*x,5*y,5*z]
```

```python
trefoil=np.array([newtrefoil(i) for i in np.arange(0,2*np.pi,0.01)])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(trefoil[:,0],trefoil[:,1],trefoil[:,2])
```
