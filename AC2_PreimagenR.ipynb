# Actividad de Clase: Funciones continuas en $\mathbb{R}$

El objetivo de esta actividad es entender como funciona la preimagen de funciones entre espacios métricos.

Por ejemplo, consideremos la función $f:[0,2\pi]\to [-1,1]$ dada por $f(x)=\sin(x).$ Queremos saber si esta función es continua cuando consideramos la topología subespacio de $[0,2\pi]$ y $[-1,1]$ cuando se consideran subconjuntos de $(\mathbb{R},d_e).$

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import plotly.express as px


X = np.linspace(0,2*np.pi,1000)

Y = np.sin(X)

fig = px.scatter(x=X, y=Y)
fig.show()
```

```python
map={x:np.sin(x) for x in X}
f= lambda x: map[x]
# We wanto to find the preimage of an open in [-1,1]
def preimage(f, lim_inf,lim_sup, Domain): # Given a function, a subset A of the codomain and the Domain,
# the preimage function will return the preimage of A.
    aux = []
    # First we look for all preimage set of each element of A
    for x in Domain:
        if f(x) >= lim_inf and f(x) <= lim_sup:
            aux.append(x)
    # Finally we return the preimage set
    return aux
```

```python
Ainv=preimage(f,-0.5,0.5,X)
Ainv
```

```python
A=[f(x) for x in Ainv]
#plt.scatter(Ainv,A
fig = px.scatter(x=Ainv, y=A)
fig.show()
```

```python
sin=pd.DataFrame({'x':X,'y':Y})
sin.head()
```

```python
flag=[x in Ainv for x in X]
flag
```

```python
sin['flag']=flag
```

```python
fig = px.scatter(x=sin["x"], y=sin["y"], color=sin["flag"])
fig.show()
```
